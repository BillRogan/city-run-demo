```python
import os

from pypdf import PageObject, PdfReader
```

```python
input_pdf_path: str = os.path.join("..", "..", "data", "raw", "city-run-ausschreibung-2026.pdf")
output_dir_path: str = os.path.join("..", "..", "data", "raw", "images")

reader: PdfReader = PdfReader(input_pdf_path)
```

```python
def export_images_from_page(page: PageObject, index: int) -> None:
    for i, image_file_object in enumerate(page.images):
        file_path = os.path.join(output_dir_path, f"page-{index}-out-image-{i}-{image_file_object.name}")
        image_file_object.image.save(file_path)

for index, page in enumerate(reader.pages):
    export_images_from_page(page, index)
```

```python
page = reader.pages[0]
xobjs = page["/Resources"]["/XObject"]

for name, ref in xobjs.items():
    obj = ref.get_object()
    print(f"--- {name} ---")
    for k, v in obj.items():
        print(k, v)
```
